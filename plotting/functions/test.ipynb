```python
import xarray as xr

# List of file paths to concatenate
file_paths = [
    "/home/grp2/dhruva-sharma/scientific-dashboard/webappfiles/data/EFD/CSES_01_EFD_1_L02_A1_272621_20221231_193728_20221231_201442_000.zarr.zip",
    "/home/grp2/dhruva-sharma/scientific-dashboard/webappfiles/data/EFD/CSES_01_EFD_1_L02_A1_272620_20221231_185008_20221231_192714_000.zarr.zip"
]

# Open each dataset
datasets = [xr.open_zarr(fp) for fp in file_paths]

# Inspect dimensions of each dataset
for i, ds in enumerate(datasets):
    print(f"Dataset {i} dimensions:", ds.dims)

# Determine the common dimensions across all datasets
common_dims = set(datasets[0].dims)
for ds in datasets[1:]:
    common_dims &= set(ds.dims)

# Drop extra dimensions from each dataset
adjusted_datasets = []
for ds in datasets:
    drop_dims = set(ds.dims) - common_dims
    adjusted_datasets.append(ds.drop_dims(drop_dims))

# Verify dimensions after adjustment
for i, ds in enumerate(adjusted_datasets):
    print(f"Adjusted Dataset {i} dimensions:", ds.dims)

# Concatenate the datasets along a common dimension, assuming 'phony_dim_0' can be used
print(adjusted_datasets)
combined = xr.concat(adjusted_datasets, dim='phony_dim_0')

# Print the combined dataset for inspection
print(combined)

# Optionally, save the combined dataset to a new NetCDF file
combined.to_zarr('/home/grp2/dhruva-sharma/scientific-dashboard/webappfiles/data/EFD/combined_dataset.zarr')
```

```python
import xarray as xr
import numpy as np

# List of file paths to concatenate
file_paths = [
    "/home/grp2/dhruva-sharma/scientific-dashboard/webappfiles/data/EFD/CSES_01_EFD_1_L02_A1_272621_20221231_193728_20221231_201442_000.zarr.zip",
    "/home/grp2/dhruva-sharma/scientific-dashboard/webappfiles/data/EFD/CSES_01_EFD_1_L02_A1_272620_20221231_185008_20221231_192714_000.zarr.zip"
]

# Open each dataset
datasets = [xr.open_zarr(fp) for fp in file_paths]

# Inspect the variables in each dataset to find the time variable
#for i, ds in enumerate(datasets):
#    print(f"Dataset {i} variables:", ds.variables)

# Assuming 'FREQ' is the correct time dimension (replace with the correct variable if different)
# Flatten the time coordinates and find unique values
flattened_times = np.concatenate([ds['FREQ'].values.flatten() for ds in datasets])
unique_times = np.unique(flattened_times)

# Convert to a DataArray for reindexing
all_times = xr.DataArray(unique_times, dims='FREQ')

# Reindex each dataset to the union of all time coordinates
datasets = [ds.reindex({'FREQ': all_times}) for ds in datasets]

# Handle mismatched dimensions by dropping 'phony_dim_0' if necessary
datasets = [ds.drop_dims('phony_dim_0') for ds in datasets]

# Concatenate the datasets along the 'FREQ' dimension
combined = xr.concat(datasets, dim='UTC_TIME')
print(combined)
# Print the combined dataset for inspection
for i, ds in enumerate(combined):
    print(ds)
    print(f"Dataset {i} variables:", ds.variables)

# Optionally, save the combined dataset to a new NetCDF file
# combined.to_netcdf('/path/to/save/combined_file.nc')
```
