```python
def find_intersection(
    file_path: str, start_date: str, end_date: str, polygon: list | Polygon
):
    if isinstance(polygon, list):
        polygon = Polygon(polygon)

    # Read the GPKG file and open the dataset
    data = gpd.read_file(file_path)
    # display(data)

    # Convert start_date and end_date to datetime objects
    start_date = datetime.strptime(start_date, "%Y-%m-%d")
    end_date = datetime.strptime(end_date, "%Y-%m-%d")

    # # Convert start_time and end_time columns to datetime
    data["start_time"] = pd.to_datetime(data["start_time"])
    data["end_time"] = pd.to_datetime(data["end_time"])

    filtered_data = data[
        (data["start_time"] >= start_date) & (data["end_time"] <= end_date)
    ]

    # Check for intersection with the given polygon and filter
    intersects = filtered_data["geometry"].intersects(polygon)
    # display(intersects)
    intersecting_data = filtered_data[intersects]
    # display(intersecting_data)

    # Return the list of semiorbit_nr that intersect with the polygon
    return intersecting_data["semiorbit_nr"].tolist()
```
